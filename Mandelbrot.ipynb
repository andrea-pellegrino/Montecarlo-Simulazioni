```python
import numpy as np
import matplotlib.pyplot as plt

def divergetest(c, vallim, itlim):
    z = c
    i = 0
    while i < itlim and np.sqrt((z.real)**2 + (z.imag)**2)< vallim:
        z = z**2 + c
        i += 1
    return i

def plotter(n, vallim, itlim, xi,xs,yi,ys):
    image = np.full((n,n),0)
    Xp = np.linspace(xi,xs,n)
    Yp = np.linspace(yi,ys,n)
    for x in range(n):
        for y in range(n):
            image[y][x] = divergetest(complex(Xp[x],Yp[y]),vallim, itlim)
    return image

n = 1000
xi,xs,yi,ys = -3,1,-2,2
vallim = 3
itlim = 30
mandelbrot = plotter(n,vallim,itlim,xi,xs,yi,ys)

plt.figure(1,figsize=[8,8])
plt.title('Mandelbrot Set')
plt.imshow(mandelbrot,extent=[-3,1,-2,2])
plt.xlabel('Real')
plt.ylabel('Complex')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def divergetest(z0, c, vallim, itlim):
    z = z0
    i = 0
    while i < itlim and np.sqrt((z.real)**2 + (z.imag)**2)< vallim:
        z = z**2 + c
        i += 1
    return i

def plotter(C, n, vallim, itlim, xi,xs,yi,ys):
    image = np.full((n,n),0)
    Xp = np.linspace(xi,xs,n)
    Yp = np.linspace(yi,ys,n)
    for x in range(n):
        for y in range(n):
            image[y][x] = divergetest(complex(Xp[x],Yp[y]),C,vallim, itlim)
    return image

C = 0.3465368236472943 + 0.07709794589178376j
n = 1000
xi,xs,yi,ys = -2,2,-2,2
vallim = 3
itlim = 30
mandelbrot = plotter(C, n,vallim,itlim,xi,xs,yi,ys)

plt.figure(1,figsize=[8,8])
plt.title('Mandelbrot Set')
plt.imshow(mandelbrot,extent=[-2,2,-2,2])
plt.xlabel('Real')
plt.ylabel('Complex')
plt.show()
```
