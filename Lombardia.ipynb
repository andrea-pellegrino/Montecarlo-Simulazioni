```python
import numpy as np
import pandas as pd

# Colours #10
# #ba2c31,#c44039,#e67b59,#fad0b9,#99bdd5,#2869ae,#2166ac

path = '/Users/andre/Documenti/Documenti_di_testo/'
file = 'Regionali.csv'
name = path + file
df = pd.read_csv(name,';')

Regione = 'Lombardia'
df["REGIONE"] = df["REGIONE"].str.title()
df["COMUNE"] = df["COMUNE"].str.title()

df = df[df["REGIONE"] == Regione]
Des = df[df["COGNOME"] == "FONTANA"]
Sin = df[df["COGNOME"] == "GORI"]
Comuni = df.COMUNE.unique()

Desp = Des.pivot(index = "COMUNE", columns = "LISTA", values = "VOTILISTA").fillna(0)
Sinp = Sin.pivot(index = "COMUNE", columns = "LISTA", values = "VOTILISTA").fillna(0)
Desp['Tot'] = Desp.sum(axis=1)
Sinp['Tot'] = Sinp.sum(axis=1)

Voti = pd.DataFrame(index = Comuni)
Voti['Comuni'] = Comuni
Voti['CDX'] = Desp['Tot']
Voti['CSX'] = Sinp['Tot']
Voti = Voti.reset_index(drop=True)

Ts = sum(Voti.CSX)
Td = sum(Voti.CDX)
OldRatio = Ts/(Ts + Td)

Voti['Total'] = Voti['CSX'] + Voti['CDX']
Voti['Ratio'] = Voti['CSX']/(Voti['Total'])

# % of Democrats
NewRatio = 0.55
flag = 1

Voti['NewRatio'] = Voti['Ratio']
Voti['Gain'] = (1-Voti['NewRatio'])
delta = 0.001

while flag == 1:
    Voti['NewRatio'] = Voti['NewRatio'] + delta*Voti['Gain']
    Voti['Gain'] = (1-Voti['NewRatio'])
    Voti['Sin'] = round(Voti['Total']*Voti['NewRatio'])
    Voti['Des'] = Voti['Total']-Voti['Sin']
    
    Td2 = sum(Voti.Des)
    Ts2 = sum(Voti.Sin)
    R = Ts2/(Td2+Ts2)
    
    if R > NewRatio:
        flag = 0

Voti['Ratio'] = (Voti['Sin']/(Voti['Sin']+Voti['Des']))
df2 = Voti[['Comuni','Ratio']]
pd.options.display.max_rows = 9999
pd.options.display.max_columns = 9999
df2 = df2.round(2)
df2 = df2.replace("u'","ù", regex=True)
df2 = df2.replace("e'","è", regex=True)
df2 = df2.replace("o'","ò", regex=True)
df2 = df2.replace(" Dè "," de' ", regex=True)
df2 = df2.replace(" Di "," di ", regex=True)
df2 = df2.replace(" Al "," al ", regex=True)
df2 = df2.replace(" Con "," con ", regex=True)
df2 = df2.replace(" Del "," del ", regex=True)
df2 = df2.replace(" Ed "," ed ", regex=True)
df2 = df2.replace(" Sul "," sul ", regex=True)
df2 = df2.replace(" Sopra "," sopra ", regex=True)
df2 = df2.replace(" Della "," della ", regex=True)
df2 = df2.replace(" Dell'"," dell'", regex=True)
df2 = df2.replace(" In "," in ", regex=True)
df2 = df2.replace(" Il "," il ", regex=True)
df2 = df2.replace(" D'"," d'", regex=True)
df2 = df2.replace(" E "," e ", regex=True)
df2 = df2.replace(" A "," a ", regex=True)
df2 = df2.replace(" Su "," su ", regex=True)
df2 = df2.replace(" Sul "," sul ", regex=True)
df2 = df2.replace(" Sull'"," sull'", regex=True)
df2 = df2.replace(" Dall'"," dall'", regex=True)
df2 = df2.replace(" Per "," per ", regex=True)
df2 = df2.replace(" Degli "," degli ", regex=True)
df2 = df2.replace("Gadesco Pieve","Gadesco-Pieve", regex=True)
df2 = df2.replace(" Dei "," dei ", regex=True)
df2 = df2.replace(" Xxiii"," XXIII", regex=True)
df2 = df2.replace(" Delle "," delle ", regex=True)
df2 = df2.replace("Puegnago del Garda","Puegnago sul Garda", regex=True)
df2 = df2.replace("Gabbioneta Binanuova","Gabbioneta-Binanuova", regex=True)
df2
```
